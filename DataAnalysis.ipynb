```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.hist(data['score'], bins=10, color='rebeccapurple', edgecolor='white')
plt.xlabel('Average Score')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Define the subject score columns
subject_columns = [
    'math_score', 'history_score', 'physics_score',
    'chemistry_score', 'biology_score', 'english_score',
    'geography_score'
]

# Calculate the average score for each student
data['score'] = data[subject_columns].mean(axis=1).round(2)

subject_averages = data[subject_columns].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
subject_averages.plot(kind='bar', color='rebeccapurple')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 6))
plt.scatter(data['weekly_self_study_hours'], data['score'], alpha=0.6, color='rebeccapurple')
plt.xlabel('Weekly Self-Study Hours')
plt.ylabel('Average Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
